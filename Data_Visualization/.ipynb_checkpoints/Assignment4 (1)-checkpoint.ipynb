```python
import pandas as pd
import matplotlib.pyplot as plt
import re
import numpy as np
import seaborn as sns

# use the 'seaborn-colorblind' style
plt.style.use('seaborn-dark')
```

```python
plt.style.available
```

```python
# get the data
mets = pd.read_csv('sportsref_mets.csv')
yanks = pd.read_csv('sportsref_download.csv')
```

```python
# check how far back the yearly data goes..
yanks_first_year, mets_first_year = yanks['Year'].min(), mets['Year'].min()

# cut back data to match years
yanks = yanks[yanks['Year'] >= mets_first_year]
yanks
```

```python
# clean up 'Manager' columns for both Yankees and Mets
yanks['Managers'] = yanks['Managers'].str.replace(r'\(.*\)', '')
mets['Managers'] = mets['Managers'].str.replace(r'\(.*\)', '')

# groupby managers and get info about them
yank_mans = yanks.groupby('Managers')['W-L%'].agg({'Avg': lambda x: x.mean(), '# Years': lambda x: x.count()})

# again for mets..
met_mans = mets.groupby('Managers')['W-L%'].agg({'Avg': lambda x: x.mean(), '# Years': lambda x: x.count()})

# plot Avg W-L% vs #Years
plt.figure()
plt.plot(met_mans['# Years'], met_mans['Avg'], 'o', color='orange', label='Mets')
plt.plot(yank_mans['# Years'], yank_mans['Avg'], 'o', color='blue', alpha=0.5, label='Yankees')

plt.legend(loc=4)
plt.title('Managers Number of Years vs. Average Win-Loss Percentage\n For the Yankees and the Mets')
plt.xlabel('# Years as Manager')
plt.ylabel('Win-Loss %')
plt.xlim(0, 13.5)
# plt.ylim(0.25, 0.75)
```

```python
%matplotlib notebook
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
ax1 = plt.subplot(111)
# ax1.plot(mets['Year'], mets['W-L%'].rolling(window=10).mean(), 
#          '-', label='Mets', color='orange', alpha=0.9)
ax1.plot(yanks['Year'], yanks['W-L%'].rolling(window=10).mean(), 
         '-', 
         label='Win_loss %', 
         color='navy', 
         alpha=0.7)

ax1.legend(loc=2)
ax1.set_ylabel('Win-Loss % (10-year running mean)')
ax1.set_title('Winning Stats of New York Major League Baseball Teams (1962-2017)')
```

```python
ax2 = ax1.twinx()
ax2.plot(yanks['Year'], yanks['Attendance'].rolling(10).mean(),
         'o', 
         color='lightblue', 
         alpha=0.9,
         label='Attendance')
# ax2.plot(mets['Year'], mets['Attendance'].rolling(10).mean(), 'o', color='darkorange', alpha=0.5)
ax2.set_ylabel('Attendance')
ax2.legend(loc=4)
```

```python
plt.subplot(222)
plt.plot(yanks['10yr Mean'], yanks['Attendance'], 'o', color='navy')
```

```python
# managers
y_managers = yanks['Managers'].unique()
y_managers = [man for man in y_managers if yanks.groupby('Managers')['Year'].count().loc[man] > 5]

m_managers = mets['Managers'].unique()
m_managers = [man for man in m_managers if mets.groupby('Managers')['Year'].count().loc[man] > 5]
```

```python
### yanks['WL Mean'] = yanks['W-L%'].rolling(window=5).mean()
yanks.head()
```

```python
plt.figure()

yanks['10yr Mean'] = yanks['W-L%'].rolling(10).mean()
mets['10yr Mean'] = mets['W-L%'].rolling(10).mean()

yanks['5yr Mean'] = yanks['W-L%'].rolling(5).mean()
mets['5yr Mean'] = mets['W-L%'].rolling(5).mean()


plt.plot(yanks['Year'], yanks['10yr Mean'], '-', label='Yankees', color='navy', alpha=0.7)
plt.plot(mets['Year'], mets['10yr Mean'], '-', label='Mets', color='orange', alpha=0.7)

blue, red = 0.0, 0.0
for yank in y_managers:
    stats = yanks[yanks['Managers'] == yank]
    last_year = stats['Year'].max()
    blue += 0.25
    plt.plot(last_year, stats['W-L%'].mean(), 'o', label=yank, color='navy')
    
for met in m_managers:
    stats = mets[mets['Managers'] == met]
    last_year = stats['Year'].max()
    red += 0.25
    plt.plot(last_year, stats['W-L%'].mean(), 'o', label=met, color='orange')

    
plt.legend()
plt.xlabel('Year')
```

In this [paper](https://www.microsoft.com/en-us/research/wp-content/uploads/2016/02/Ferreira_Fisher_Sample_Oriented_Tasks.pdf) the authors describe the challenges users face when trying to make judgements about probabilistic data generated through samples. As an example, they look at a bar chart of four years of data (replicated below in Figure 1). Each year has a y-axis value, which is derived from a sample of a larger dataset. For instance, the first value might be the number votes in a given district or riding for 1992, with the average being around 33,000. On top of this is plotted the 95% confidence interval for the mean (see the boxplot lectures for more information, and the yerr parameter of barcharts).

A challenge that users face is that, for a given y-axis value (e.g. 42,000), it is difficult to know which x-axis values are most likely to be representative, because the confidence levels overlap and their distributions are different (the lengths of the confidence interval bars are unequal). One of the solutions the authors propose for this problem (Figure 2c) is to allow users to indicate the y-axis value of interest (e.g. 42,000) and then draw a horizontal line and color bars based on this value. So bars might be colored red if they are definitely above this value (given the confidence interval), blue if they are definitely below this value, or white if they contain this value.

```python
%matplotlib notebook
# Use the following data for this assignment:

import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import scipy.stats as st

np.random.seed(12345)

df = pd.DataFrame([np.random.normal(32000,200000,3650), 
                   np.random.normal(43000,100000,3650), 
                   np.random.normal(43500,140000,3650), 
                   np.random.normal(48000,70000,3650)], 
                  index=[1992,1993,1994,1995])

df
```

```python
# transpose df to access the years as columns. 
df = df.T  
```

```python
df.describe()
```

```python
'''
Get mean and 95% confidence interval for each year. 
'''
# get 95% confidence interval using scipy.stats.t.interval() 
def get_confidence(data, percent=0.95):
    a = np.array(data)  # get the entire sample as array
    result = st.t.interval(percent, len(a)-1, loc=np.mean(a), scale=st.sem(a))
    return result

confidence = df.apply(get_confidence)  # return lower and upper limit
```

```python
lower, upper = zip(*list(confidence))  # get 95% range as two lists
```

```python
means = tuple(df.apply(np.mean))  # get bar heights as tuple
std = tuple(df.apply(np.std))     # get y_error as tuple

n = len(means)                    # get num of bars (4)
ind = np.arange(n)                # set x location for the bars (indexing)
bar_width = 0.45                  # width of the bars
x_labels = list(df.columns)       # get years for x_labels as a list

# calculate y-error bars from 95% confidence interval 
y_r = [(means[i]-confidence.iloc[i][1]) for i in range(len(confidence))]

# set some color values
norm = mpl.colors.Normalize(-1, 1)
```

```python
def color_picker(bars, y):
    '''
    Takes in a matplotlib bar chart and y-axis position of a line 
    and manipulates the colors of the bar chart depending on the position of the line
    relative to the mean value of each bar.
    
    return: None
    '''
    color_nums = range(9)

    for i, bar in enumerate(bars):
        mean, conf = means[i], y_r[i]
        pdiff = (y-mean) / mean
        pdiff_conf = (mean-conf) / mean
        
        value = (1-(abs(pdiff)/2))
        
        red = value if (pdiff<0) else 0.1
        blue = value if (pdiff>0) else 0.2
        green = 0.0
        
        if (y > mean and y < mean-conf) or (y < mean and y > mean+conf):
            green = (1-abs(pdiff))/2
            
        bar.set_color((red, green, blue)) 
```

```python
# Add Interactivity w/ y-axis line created when graph is clicked
class LinePicker(object):
    def __init__(self, ax, bars):
        self.fig = ax.figure
        self.ax = ax
        self.bars = bars
        
        # set a default line to appear
        self.horizontal_line = ax.axhline(y=4.2e4, color='y', alpha=0.5)
        
        # set default text
        self.text = ax.text(4.02, 4.2e4, "42000",    # x, y, label
                            bbox=dict(facecolor='w',    # add border box
                                      alpha=0.9))
        
        print(self.horizontal_line)
        
        # call self.onclick method and tell mpl to pass it a 'button_press_event'
        self.fig.canvas.mpl_connect('button_press_event', self.onclick)
        
    def onclick(self, event):
        if event.inaxes == self.ax:                     # if the plot is clicked
            x = event.xdata                             # get event's x position
            y = event.ydata                             # get event's y position
            xlim0, xlim1 = ax.get_xlim()                # get the x-axis limits (upper and lower)
            if xlim1 >= x >= xlim0:                     # check that click happened inside the plot
                # set and draw the line
                self.horizontal_line.set_ydata(y)
                self.text.set_text(str(int(y)))         # add the y-position as text label
                self.text.set_position((xlim0-0.5, y))  # place the text to the right
                self.fig.canvas.draw()                  # draw it
            # set bar colors based on y-position of line
            color_picker(bars, y)
                
```

```python
# create the figure and a LinePicker instance on it                
if __name__ == '__main__':
    fig = plt.figure()
    ax = fig.add_subplot(111)
    
    bars = ax.bar(left=ind,                         # position of bars
        height=means,                               # bar heights
        width=bar_width,                            # bar widths
        yerr=y_r,                                   # error bars
        capsize=3,                                  # perpendicular caps on error bars (default=None)
        tick_label=x_labels,                        # add x-axis labels
        bottom=0,                                   # set bottom to zero
        alpha=0.9,                                  # set opacity
        align='center') 
        
    p = LinePicker(ax, bars)
    
    # call the color_picker func to set starting colors based on starting line
    _, ya = p.horizontal_line.get_ydata()
    color_picker(bars, 42000)
    
    
    plt.show()              
    
```
